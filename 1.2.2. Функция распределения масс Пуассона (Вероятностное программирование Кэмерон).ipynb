1.3.2.  Непрерывный случай - функция плотности вероятности

```python
Рассмотрим непрерывную переменную X. Когда х пробегает значения от 0 до 1, 
вероятности накапливаются непрерывно-, таких крупных «вкладов» в общую сумму, 
как могла внести точка в случае дискретной переменной, здесь не будет. 
В непрерывном случае нельзя, указав точку х, спросить, какова вероятность того, 
что именно она окажется интересующим нас частным значением. 
Оправданным будет вопрос о скорости аккумуляции вероятностей для каждого значения х. 
(Здесь может быть полезна аналогия: в какой бы точке ни находился движущийся автомобиль, 
неправомерно спрашивать, как далеко он уехал в данный момент; 
правильнее задать вопрос о скорости, с которой автомобиль в данный момент движется.) 
Скорость, с которой аккумулируется вероятность, задает тангенс угла наклона графика 
функции распределения к оси абсцисс: чем круче его наклон, тем быстрее идет 
накопление вероятностей в функции распределения. 

Теперь мы можем определить новую функцию, которую будем обозначать fx (•):

Пусть FX(x): = P(X≤x) – функция распределения вероятности сл.в. X. 
Обозначим Δx := x-xo приращение аргумента

ΔFX(xo) = P(X≤ xo+ Δx) - P(X≤ xo) – приращение функции распределения вероятности в точке xo.

Отношение ΔFX(xo)/ Δx = { P(X≤ xo+ Δx)- P(X≤ xo)}/ Δx – можно назвать 
плотностью распределения вероятности на интервале, 
а если мы устремим интервал Δx к нулю, 
то получим скорость изменения плотности распределения вероятности в точке xo:

fX(xo) =lim(∆x→0)⁡  { P(X≤ xo+ Δx)- P(X≤ xo) }/ Δx  =: F’X(x) = dFX(x)/dx
или, по-другому, плотность вероятности в точке xo.
Эту функцию и договорились называть функцией плотности вероятности сл. величины X.
У нас говорят случайная величина, а у них - случайная переменная или стохастическая переменная.

Чем шире график плотности распределения, тем есть степень уверенности ниже;
Чем распределение уже, тем степень нашей уверенности выше.
```

Экспоненциальное распределение часто описывает время до наступления события. 

```python
%matplotlib inline
from IPython.core.pylabtools import figsize 
import numpy as np
from matplotlib import pyplot as plt
import scipy.stats as stats

figsize(12.5, 4)

a = np.linspace(0, 4, 100)
expo = stats.expon
alpha_ = [0.5, 1, 5, 10] # среднее число событий за единицу времени
colors = ['#348ABD','#A60628','#bca534','#665798']
for l, c in zip(alpha_, colors):
    plt.plot(a, expo.pdf(a, scale=1./l), lw=3,
             color=c, label=chr(945)+"= %.1f" % l)
    plt.fill_between(a, expo.pdf(a, scale=1./l), color=c, alpha=.33)
plt.legend()
plt.ylabel("Функция плотности вероятности в точке $z$")
plt.xlabel("$z$")
plt.ylim(0,1.2)
plt.xlim(0, 4)
plt.title("Функция плотности вероятности экспоненциальной случайной переменной\n при различных значениях "+chr(945))
```

```python
%matplotlib inline
from IPython.core.pylabtools import figsize 
import numpy as np
from matplotlib import pyplot as plt
import scipy.stats as stats

figsize(12.5, 4)

a = np.linspace(0, 4, 100)

def expon (z, alfa): # fZ(z|α) = α × exp(–αz), z ≥ 0
    return (1./alfa) * np.exp(- (1./alfa) * z)

alpha_ = [0.5, 1, 5, 10] # среднее число событий за единицу времени
colors = ['#348ABD','#A60628','#bca534','#665798']
for l, c in zip(alpha_, colors):
    plt.plot(a, expon(a, 1./l), lw=3,
             color=c, label=chr(945)+"= %.1f" % l)
    plt.fill_between(a, expon(a, 1./l), color=c, alpha=.33)
plt.legend()
plt.ylabel("Функция плотности вероятности в точке $z$")
plt.xlabel("$z$")
plt.ylim(0,1.2)
plt.xlim(0, 4)
plt.title("Функция плотности вероятности экспоненциальной случайной переменной \n при различных значениях "+chr(945))
```
