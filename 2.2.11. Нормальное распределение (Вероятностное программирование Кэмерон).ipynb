```python
%matplotlib inline
from IPython.core.pylabtools import figsize 
import numpy as np
from matplotlib import pyplot as plt
import scipy.stats as stats

figsize(14.5, 4)

def norm_gauss(x, mu = 0, tau = 1):
    return (tau/(2*np.pi))**(1/2) * np.exp(np.dot(-tau/2, (x - mu)**2)) 
x = np.linspace(-10, 10, 500)

plt.plot(x, norm_gauss(x), label=r"$\mu = 1$", ls="-", lw=1)
plt.vlines([-1,1], 0, .5, label=r"$\mu+/-\sigma$", color='#A60628', ls="--", alpha=0.7)
plt.vlines([-2,2], 0, .5, label=r"$\mu+/-2\sigma$", color='#a9df20', ls="--", alpha=0.7)
plt.vlines([-3,3], 0, .5, label=r"$\mu+/-3\sigma$", color='#20b3df', ls="--", alpha=0.7)

plt.title(u"Функция плотности стандартного нормального распределения при $\\mu = 0$ и $\\tau = 1$")
plt.legend(loc="upper left")
plt.show()
# -----------------------график 2-й------------------------------
plt.plot(x, norm_gauss(x, 1), label=r"$\mu = 1$", ls="--", lw=1)
plt.plot(x, norm_gauss(x, 3), label=r"$\mu = 3$", ls="--", lw=1)
plt.plot(x, norm_gauss(x, -5), label=r"$\mu = -5$", ls="--", lw=1)
plt.legend(loc="upper left")
plt.show()
# -----------------------график 3-й---------------------------
plt.plot(x, norm_gauss(x, 1, 1), label=r"$\mu = 1, \tau = 1$", color="#348ABD")
plt.fill_between(x, norm_gauss(x, 1, 1), color="#348ABD",alpha=.33)
plt.plot(x, norm_gauss(x, 1, 3), label=r"$\mu = 1, \tau = 3$", color="#A60628")
plt.fill_between(x, norm_gauss(x, 1, 3), color="#A60628",alpha=.33)
plt.plot(x, norm_gauss(x, 1, 10), label=r"$\mu = 1, \tau = 10$", color="#7A68A6")
plt.fill_between(x, norm_gauss(x, 1, 10), color="#7A68A6",alpha=.33)
plt.title(u"Чем меньше $\\tau$, тем график нормального распределения шире")
plt.xlabel("$x$")
plt.ylabel(u"Нормальная плотность от $x$")
plt.xlim(-4,6)
plt.legend(loc="upper left")
plt.show()
```

```python
import numpy as np
def norm_gauss(x, mu = 0, tau = 1):
    return (tau/(2*np.pi))**(1/2) * np.exp(np.dot(-tau/2, (x - mu)**2)) 

def gauss (x0 = -5, x1 = 5):
    step_ = 100000
    y =  np.linspace(x0, x1, step_, retstep=True)
    return np.sum(norm_gauss(y[0])*y[1])# здесь y[0] - массив данных, y[1] - шаг

def gauss1 (x1 = 5):
    x0 = -5
    step_ = 100000
    y =  np.linspace(x0, x1, step_, retstep=True)
    return np.sum(norm_gauss(y[0])*y[1])
```

```python
XA = np.linspace(-5, 5)

for i in XA:
    print('P[X< {0:.2}] = {1:.2%}'.format(i, gauss(-5,i)))
```

Построим несколько графиков различных функций плотности нормального 
распределения

```python
import scipy.stats as stats

nor = stats.norm
x = np.linspace(-8, 7, 500)
mu = (-2, 0, 3)
tau = (.7, 1, 2.8)
colors = ["#348ABD", "#A60628", "#dfb120"] # hex-коды цветов можно взять https://color2.ru/
parameters = zip(mu, tau, colors)

for _mu, _tau, _color in parameters:
    plt.plot(x, nor.pdf(x, _mu, scale=1./_tau),
             label="$\mu = %d,\;\\tau = %.1f$" % (_mu, _tau),
             color=_color)
    plt.fill_between(x, nor.pdf(x, _mu, scale=1./_tau), color=_color, alpha=.33)

plt.legend(loc="upper right")
plt.xlabel("$x$")
plt.ylabel(u"Функция плотности от $x$")
plt.title(u"Распределение плотности вероятности трех различных нормальных случайных переменных")    
```
