```python
class Student: # базовый класс
    """DB students"""
    
    def __init__(self, name):
        self.name = name
    def show(self):
        print('ФИО: \t'+self.name)
    def random(self):
        print('Некий метод базового класса')

class RegularStudent(Student): #производный класс
    
    def __init__(self,name): # переопределяем метод базового класса 
        Student.__init__(self,name) # вызываем метод базового класса
        self.age = 22 # дополнительный член класса
    
    def show(self): # переопределеяем метод базового класса
        print('ФИО из базового класса: '+self.name+', \nВозраст - из производного: '+str(self.age))
    
    def random(self): # переопределяет метод базового класса
        Student.random(self) # вызывает метод базового класса
        print('++'*25)
        print('Некий метод в производном классе')
```

```python
class Student:
    def __init__(self, name, email):
        self.name = name # член данных
        self.email = email  # член данных
        
    def putdata(self):
        print('=='*25)
        print('Данные Студента: \n ФИО: ',self.name, 
              '\n E-mail: ', self.email)
class PhDGuide:
    def __init__(self, name, email, students): # здесь students - это список
        self.name = name
        self.email = email
        self.students = students
    
    def putdata(self):
        print('=='*25)
        print('Данные Наставника: \n ФИО: ',self.name, 
              '\n E-mail: ', self.email)
        print('\n Список подопечных студентов:')
        
        if self.students :
            for s in self.students:
                print(s.name, s.email)
        else:
            print('Учеников нет')
    
    def add(self, student):
        s = Student(student.name, student.email)
        if s not in self.students:
            self.students.append(s)
            print('\n'+'**'*25)
            print('Студент: ',s.name, s.email, 'назначен учеником', self.name)
            print('**'*25+'\n')
            
    def remove(self, student):
        s = Student(student.name, student.email)
        flag = 0
        for s1 in self.students:
            if s1.email == s.email:
                print('\n'+'**'*25)
                print('Студент: ',s.name, s.email, 'исключён из списка учеников', self.name)
                print('**'*25+'\n')
                self.students.remove(s1)
                flag = 1
        if flag == 0:
            print('Не найден')
        
```

11.2. НАСЛЕДОВАНИЕ: ВАЖНОСТЬ И ТИПЫ

```python
class Book:
    def getdata(self):
        self.name = input('Введите название книги: ')
        self.n = int(input('Введите количество авторов: '))
        self.authors = []
        i = 0
        while i < self.n:
            author = input(str('Введите ФИО '+str(i+1)+'-го автора'))
            self.authors.append(author)
            i += 1
        self.publisher=input("\n Введите издателя: ")
        self.ISBN=input("\nВведите ISBN: ")
        self.year=input("\nВведите год публикации: ") 
    
    def putdata(self):
        print("\nНазвание: ", self.name,
              "\nАвтор(ы): ", self.authors,
              "\nИздатель: ", self.publisher,
              "\nГод: ", self.year,
              "\nISBN: ", self.ISBN)

class Text_Book:
    def getdata(self):
        self.course = input('\n Введите курс: ')
        Book.getdata(self)
    def putdata(self):
        Book.putdata(self)
        print('\n Курс: ', self.course)
```

```python
Book1=Book()
Book1.getdata()
Book1.putdata()
TextBook1 = Text_Book() 
TextBook1.getdata() 
TextBook1.putdata()
```

11.3.5. Вызов функции базового класса с помощью Super

```python
class BaseClass:
    def __init__(self, data):
        self.data = data
    def printData(self):
        print('Обработка данных методом базового класса:', self.data)
        
class DerivedClass(BaseClass):
    def __init__(self, data1, data2):
        self.data1 = data1
        super(DerivedClass, self).__init__(data2) # передали данные в базовый метод __init__
    def printData(self):
        super(DerivedClass, self).printData() # обработали данные другим базовым методом
        print('Обработка данных методом производного класса:', self.data1)
        
```

```python
d1 = DerivedClass(4,5)
d1.printData()
```

11.4. ПОИСК В ДЕРЕВЕ НАСЛЕДОВАНИЯ

```python
class BaseClass:
    def __init__(self, data):
        self.data = data
    def show(self):
        print('\n Данные обработаны базовым классом: ', self.data)
        
class Derived1(BaseClass):
    def __init__(self, data, data1):
        self.data1 = data1
        BaseClass.__init__(self, data)
    def show(self):
        print('\n Данные обработаны производным1 классом',
             self.data1,
             '\n Данные обработаны базовым классом',
             self.data)
        
class Derived2(Derived1):
    def __init__(self, data0, data1, data2):
        self.data0 = data0
        #BaseClass.__init__(self, data0)
        Derived1.__init__(self, data1, data2)
        
    def show(self):
        print('\n Данные обработаны производным2 классом', self.data0)
        Derived1.show(self)
        BaseClass.show(self)

X = Derived2(10,4,8)

X.show()
```

# 11.5. ИНТЕРФЕЙС КЛАССА И АБСТРАКТНЫЕ КЛАССЫ

```python
class BaseClass:
    def method1(self):
        print("In BaseClass from method1") 
    def method2(self):
        self.action()
        print('Метод Action выполнен')

class Derived1(BaseClass): 
    def method1(self): # базовый метод переопределен
        print("A new method, has got nothing to do with that of the base class") 

class Derived2(BaseClass):
    def method1(self): # базовый метод дополнен
        print("A method that extends the base class method")
        BaseClass.method1(self) 

class Derived3(BaseClass): 
    def action(self):
        print("\nЗдесь запустили action-метод базового класса")

for className in (Derived1, Derived2, Derived3):

    print("\nClass\t:",className)
    className().method1()
    print('--'*25)

print('=='*25)    
X = Derived3()
X.method2()
```
