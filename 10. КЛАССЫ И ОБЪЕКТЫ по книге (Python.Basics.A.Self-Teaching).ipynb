10.4. ОБЛАСТЬ ВИДИМОСТИ ДАННЫХ

```python
class demo_class:
    a = 25 # переменная класса видна только в первом методе, но в нём она переопределена
    f = 38
    def getdata(self, b, a=a):
        # a = 7 # а - локальная переменная getdata, т.е. в других методах она не видна
        self.a = a # член данных self.* класса - виден во всех методах класса
        self.b = b # член данных класса
    
    def putdata(self):
        
        print('self.b =', self.b)
        print('self.a =', self.a)
        print('f =', f)
        print('a =', a) # а - локальная переменная putdata - видна  только внутри putdata

d = demo_class() #создаётся экземпляр класса - объект класса

d.getdata(9)
print('d.a =', d.a)
print('d.b =',d.b)
print('d.f =',d.f)
d.putdata()
```

```python
class demo_class1:
    a = 5 # переменная класса - это переменная, общая для всех экземпляров
    def getdata(self,b):
        c = 7
        self.b = b # элементу данных присвоено значение b
        print('c =', c, ' и self.b = ', self.b)
        
    def other_function(self):
        c = 3
        print('c =', c)
    def putdata(self):
        print('self.b = ',self.b)

d1 = demo_class1()
print('объект d1')
d1.getdata(10)
print('d1.a =',d1.a)
d1.other_function()
d1.putdata()
print('=='*20)
print('объект e')
e = demo_class1()
print('e.a =',e.a)
e.other_function()
e.putdata()
```

```python
global f
f = 7 #глобальная переменная

class demo_class2:
    a = 5 # переменная класса
    def getdata(self, b):
        c = 7
        self.b = b # член данных класса
        print('c =', c, 'self.b =', self.b, 'f =', f)
    def other_function(self):
        c = 3 # локальная переменная
        print('c =', c)
    def putdata(self):
        print('self.b =', self.b)

d2 = demo_class2()
d2.getdata(13)
print('d2.a =', d2.a)
d2.other_function()
d2.putdata()

e2 = demo_class2()
print('e2.a =', e2.a)
```

10.5. ВЛОЖЕНИЕ

```python
class date:
    def getdata(self):
        self.dd = input('Введите день (dd) \t:')
        self.mm = input('Введите месяц (mm) \t:')
        self.yy = input('Введите год (yy)\t:')
    def display(self):
        print('Дата рождения: '+self.dd+':'+self.mm+':'+self.yy)
        
class student:
    def getdata(self):
        self.name = input('Введите ФИО \t:')
        self.dob = date() # создаём экземпляр внешнего класса date
        self.dob.getdata() # используем метод1 объекта
    def putdata(self):
        print('ФИО: ', self.name)
        self.dob.display()# используем метод2 объекта
        
s = student()
s.getdata()
s.putdata()
```

```python
class movie:
    def getdata(self):
        self.name = input('Введите название фильма \t:')
        self.year = int(input('Введите год \t:'))
        self.genre = input('Введите жанр фильма \t:')
        self.director = input('Введите режисера фильма \t:')
        self.producer = input('Введите продюсера фильма \t:')
        
        List_actors = []
        item = input('Введите ФИО актёра \t:')
        List_actors.append(item)
        
        choice = input('Нажмите \'y\' для дальнейшего ввода или  \'n\' для выхода')
        while choice == 'y':
            item = input('Введите ФИО актёра \t:')
            List_actors.append(item)
            choice = input('Нажмите \'y\' для дальнейшего ввода или  \'n\' для выхода')
        
        self.actors = List_actors
        self.music_director = input('Введите ФИО композитора \t:')
    
    def putdata(self):
        print('Фильм \t:', self.name)
        print('Год \t',self.year)
        print('Жанр \t:',self.genre)
        print('Режиссер \t:',self.director)
        print('Продюсер \t:', self.producer)
        print('Композитор \t:', self.music_director)
        print('Актёры \t:',self.actors)
    
    def __init__(self):
        self.name='Fault'
        self.year=2015
        self.genre='Drama'
        self.director='XYZ'
        self.producer='ABC'
        self.music_director='LMN'
        self.actors=['Al', 'A2', 'A3','A4']
        
# ===========конец=описания=класа=======================

m = movie()
#m.getdata()
m.putdata()
```

10.8. ДЕСТРУКТОРЫ

```python
class employee3:
    def getdata(self):
        self.name = input('Введите ФИО: \t')
        self.age = input('Введите возраст: \t')
    def putdata(self):
        print('ФИО: \t', self.name)
        print('Возраст: \t', self.age)
    def __init__(self, name = 'dima', age = 50):
        self.name = name
        self.age = age
    def __del__(self):
        print('Объект удален из памяти')
       

em31 = employee3('asdasdasd', 45)
em31.getdata()
print('=='*20)
em31.putdata()
print('=='*20)
del em31
```

```python
class employee4:
    """ The employee class"""
    
    def getdata(self):
        self.name = input('Введите ФИО: \t')
        self.age = input('Введите возраст: \t')
    def putdata(self):
        print('ФИО: \t', self.name)
        print('Возраст: \t', self.age)
    def __init__(self, name = 'dima', age = 50):
        self.name = name
        self.age = age
    def __del__(self):
        print(__class__.__name__,'удален из памяти')
       

em41 = employee4('asdasdasd', 45)
em41.getdata()
print('=='*20)
em41.putdata()
print('=='*20)
print('Строка документации em41.__doc__:',em41.__doc__)
del em41
```
